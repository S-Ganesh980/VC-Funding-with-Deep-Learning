# VC funding with Deep Learning 

Here we will predict whether an applicant will be successful if funded by a VC-firm. 
<br>
 Our dataset consists of over 34,000 organizations that are funded by a VC-firm. We will use Machine learning and neural networks to create a binary classifier model that will predict whether an applicant will be successful. 

```python
#Imports 
import pandas as pd
from pathlib import Path
import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
#Data
df = pd.read_csv('vc_funding.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#dropping EIN and NAME columns
df = df.drop(['EIN','NAME'], axis=1)
df.head()
```

```python
#Encoding the categorical variables using OneHotEncoder

cat_var = list(df.dtypes[df.dtypes=='object'].index)
```

```python
cat_var
```

```python
enc = OneHotEncoder(sparse=False)
```

```python
encoded_data = enc.fit_transform(df[cat_var])
```

```python
#creating a dataframe with encoded categorical variables

encoded_df= pd.DataFrame(encoded_data, columns=enc.get_feature_names_out(cat_var) )
```

```python
encoded_df
```

```python
#concatenating numerical data with encoded_df 

encoded_df= pd.concat([df.drop(columns= cat_var), encoded_df], axis=1)
```

```python
encoded_df.head()
```

```python
#creating matrix of features(X) and dependent variable vector(y)

X= encoded_df.drop(['IS_SUCCESSFUL'], axis=1)
X.head()
```

```python
y = encoded_df['IS_SUCCESSFUL']
```

```python
y.head()
```

```python
# splitting the dataset into training and test set

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size= 0.3, random_state=1)
```

```python
#Using StandardScaler to scale the features

scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

#fit caler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Binary Classification Model using a Neural Network 

We will create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras. We will start with a two-layer deep neural network model that uses relu activation function for both layers. 

```python
#Defining the number of input features to the model
input_features = len(X_train.iloc[0])
```

```python
input_features
```

```python
#Defining number of hidden nodes for the first hidden layer 
hidden_nodes_layer1= 8
```

```python
#Defining number of hidden nodes for the second hidden layer 
hidden_nodes_layer2= 8
```

```python
hidden_nodes_layer2
```

```python
#creating Sequential model instance
nn = Sequential()
```

```python
#Adding first hidden layer 
nn.add(Dense(units = hidden_nodes_layer1, input_dim = input_features, activation = "relu"))
```

```python
#Adding second hidden layer 
nn.add(Dense(units = hidden_nodes_layer2, input_dim = input_features, activation = "relu"))
```

```python
# Define the number of neurons in the output layer
output_neurons = 1
```

```python
# Adding output layer to the model specifying the number of output neurons and activation function
nn.add(Dense(units=output_neurons, activation = "linear"))
```

```python
#Sequential model summary
nn.summary()
```

## Compiling and fitting the model

```python
#compile sequential model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

```python
# Fit the model using 50 epochs and the training data
model = nn.fit(X_train_scaled,y_train,epochs=50)
```

## Evaluating the model's loss and accuracy

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose = 2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
